```python
"""örnek bir öğrenci oluşturdum bu öğrenciyi kullanarak diğer öğrenci profillerini oluşturdum 
ve kullanıcıdan bu değerleri girmesini istedim"""

student={
        " student name":"Fadıl",
        "midterm grade":10,
        "project grade":10,
        "final grade":10
       
}
student1={}     #5 tane öğrenci oluşturdum
student2={}
student3={}
student4={}
student5={}
listofmean=[]     #burda öğrencilerin ortalamalarını depoluyacam alttaki ise geçici olarak kullanacağım bir dict
temp={}
listofstudent=[student1,student2,student3,student4,student5]

for x in range(0,len(listofstudent)):#öğrencilerin genel taslağını oluşturdum
    listofstudent[x]=student.copy()
for x in range(0,len(listofstudent)):   #kullanıcıdan öğrencinin verilerini girmesini istedim    
        temp=listofstudent[x].copy()    #oluşturduğum bir sözlüğün içine sırasıyla öğrencinin bilgilerini alıyorum
        for k,v in temp.items():       
          print("please enter",k)
          new=input()
          v=new
          temp[k]=new
        listofstudent[x]=temp.copy()    # ardından tekrar orjinal sözlüğe aktarıyorum.
        
for x in range(0,len(listofstudent)):     # her bir öğrencinin not ortalamasını hesaplayıp listenin içine attım.
    listofmean.append(float(listofstudent[x]["midterm grade"])*0.3+float(listofstudent[x]["project grade"])*0.3
                      +float(listofstudent[x]["final grade"])*0.4)
for x in range(0,len(listofmean)):      #her bir öğrencinin ortlamasını kendi için oluşturulan sözlüğün içine aktardım.
    listofstudent[x]["mean"]=listofmean[x]
listofmean.sort()                       #burda öğrenci ortalamaları küçükten büyüğe doğru sıralandı.
listofmean.reverse()                    #bu satırda ise büyükten küçüğe doğru sıraladım.
print(listofmean)
    
for x in range(0,len(listofstudent)):   #daha uygun gözükmesi için tek tek bastırdım.
    print(listofstudent[x])
    
```
